* **Carga y Preprocesamiento de Datos:**
Carga los datos del archivo CSV utilizando NumPy.
Realiza un preprocesamiento básico para asegurarte de que los datos estén limpios y listos para su análisis.

```python
import numpy as np

'''
# Definimmos primero una lista de tipos de variables, que nos permitirá codificr adecuadamente los valores.
Vamos además a importar solo las columnas que nos interesan para el análisis
'''
dtype = [
    ("Date", "U10"),                 # string para la fecha (hasta 10 caracteres)
    ("Product Category", "U11"),     # string para categoría
    ("Quantity", "i4"),              # int
    ("Total Amount", "i4")           # int
]

#utilizamos usecols para seleccionar solo las columnas escogidas
data = np.genfromtxt("retail_sales_dataset.csv", delimiter=",", dtype=dtype, usecols=(1, 5,6,8), skip_header=1)

# tomamos las fechas, para ordenar el arreglo. A partir de esto 
fechas = np.array(data["Date"])
indices_ordenados = np.argsort(fechas)
data_ordenado = data[indices_ordenados]
print(data_ordenado)
```

podemos ver que las fechas van desde el inicio de 2023 hasta el 1 de enero de 2024, por lo que se puede interpretar que se trata de las ventas durante un año.

Podemos observar también que hay fechas en las que se produce más de una venta del mismo tipo de artículo, por lo que para calcular los promedios deberemmos tener en consideración este aspecto.

* **Exploración de Datos:**
Calcula el total de ventas por categoría de producto.
Calcula el promedio de ventas diarias por categoría de producto.
Identifica las categorías de productos con mayores y menores ventas.

```python
#detectamos primero las categorías de productos
categorias = np.unique(data['Product_Category'])
'''
con un ciclo, filtraremos los datos del dataset por categoría de producto, sumando la cantidad de ventas.
Vamos a proceder considerando la cantidad de artículos vendidos, por una parte, y el monto en dinero por otro.

Por otra parte, calcularemos promedios considerando como total la cntidad de días en que efectivaente se 
realizan ventas, por una parte, y por otra considerando todo el año, con 365 dias

'''
resumen=[]
for t in categorias:
  filtro=data['Product_Category']==t
  total_articulos=np.sum(data['Quantity'][filtro]) #suma de los artículos vendidos
  total_ingresos=np.sum(data['Total_Amount'][filtro]) #suma de las ventas, en monto

#calculamos los promedios, como se detalló

  promedio_articulos_dias=total_articulos/len(np.unique(data['Date'][filtro]))
  promedio_articulos_anio=total_articulos/365
  promedio_ingresos_dias=total_ingresos/len(np.unique(data['Date'][filtro]))
  promedio_ingresos_anio=total_ingresos/365

  q=np.array([total_articulos,total_ingresos])
  print(f'Categoría: {t}\n')
  print(f'Total de artículos vendidos: {total_articulos}')
  print(f'Total de ingresos: {total_ingresos}')
  print(f'Promedio de artículos vendidos por día efectivo: {promedio_articulos_dias:.2f}')
  print(f'Promedio de artículos vendidos por dia, general: {promedio_articulos_anio:.2f}\n')
  resumen.append(q) #guardamos los arreglos en una lista
resumen=np.array(resumen)#convertimos la lista en array
max_ventas=np.argmax(resumen[:,0])
min_ventas=np.argmin(resumen[:,0])
max_ingresos=np.argmax(resumen[:,1])
min_ingresos=np.argmin(resumen[:,1])

print(f'Categoría con más ventas: {categorias[max_ventas]}')
print(f'Categoría con menos ventas: {categorias[min_ventas]}')
print(f'Categoría con más ingresos: {categorias[max_ingresos]}')
print(f'Categoría con menos ingresos: {categorias[min_ingresos]}')
```
