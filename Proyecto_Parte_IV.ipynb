* Utiliza histogramas y boxplots para visualizar la distribución de las variables numéricas clave.

Tomaremos las variables asociadas al total pagado (Total) y la cantidad de artículos por venta (Quantity). Las demás variables se desprenden directamente de estas, por lo que son menos esenciales.

```python
import matplotlib.pyplot as plt
import seaborn as sns


def graficos_conjuntos(data, titulo):
  # Crear una figura con 2 subplots
  fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
  #Primer subplot
  ax1.hist(data, bins=5, color='skyblue', edgecolor='black', alpha=0.7)
  ax1.set_xlabel("Intervalos")
  ax1.set_ylabel("Frecuencia")
  #ax1.set_grid(True)
  ax1.legend()

  #Segundo subplot
  ax2.boxplot(data, patch_artist=True, boxprops=dict(facecolor='skyblue', color='black'), medianprops=dict(color='red'), whiskerprops=dict(color='black'), capprops=dict(color='black'))
  ax2.set_ylabel("Valores")
  #ax2.set_grid(True)
  ax2.legend()

  # Título centrado para toda la figura
  fig.suptitle(titulo, fontweight='bold', fontsize=16, y=1.05)

  # Ajustar el diseño
  plt.tight_layout()
  plt.show()

graficos_conjuntos(df['Total'], "Distribución por monto total de compra")

graficos_conjuntos(df['Quantity'], "Distribución por cantidad de artículos por compra")
```

Es posible observar que, en el caso de la distribución por monto total de la compra, la cantidad de compras va disminuyendo drásticamente a a medida que el monto de la compra aumenta. Esto se refleja en el boxplot, en el que además apreciamos la existencia de valores atípicos y otros valores altos, que, sin llegar a ser atípicos, "estiran" el gráfico hacia la parte superior.

La variación para la cantidad de artículos por compra presenta una evidente variación menor, influída por cierto por un menor rango de las mismas. En el boxplot, podemos obervar que su distribucuón es bastante uniforme, apenas levemente tendiente a valores superiores.

* Crea gráficos de líneas para mostrar las tendencias de ventas a lo largo del tiempo.

Vamos a crear los gráficos considerando por cierto la columna 'Date', y su relación con los montos totales de las ventas

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Total', color='skyblue', errorbar=None)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

No se aprecia un patrón de variación de las ventas en el tiempo.

Utilizamos el comando **pairplot()** para graficar todas las variables, de a pares, para analizar posibles correlaciones.

```python
sns.pairplot(df)
```

Es posible observar que, salvo para las variabes que ex profeso estań relacionadas una con otra, no existen correlaciones apreciables entre los pares de variables. Construiremos el diagrama de puntos para las variables Total y Rating

```python
plt.figure(figsize=(12, 6))
plt.scatter(df['Total'], df['Rating'], color='skyblue', alpha=0.5)
plt.xlabel('gross income')
plt.ylabel('Rating')
plt.title('Relación entre Total y Rating')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()

```
