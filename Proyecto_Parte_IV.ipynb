* Utiliza histogramas y boxplots para visualizar la distribución de las variables numéricas clave.

Tomaremos las variables asociadas al total pagado (Total Amount) y la cantidad de artículos por venta (Quantity). Las demás variables se desprenden directamente de estas, por lo que son menos esenciales. (la relación entre estas variables determina la relación que pueden darse entre las demás)

```python
import matplotlib.pyplot as plt
import seaborn as sns


def graficos_conjuntos(data, intervalos, titulo):
  # Crear una figura con 2 subplots
  fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
  #Primer subplot
  ax1.hist(data, bins=intervalos, color='skyblue', edgecolor='black', alpha=0.7)
  ax1.set_xlabel("Intervalos")
  ax1.set_ylabel("Frecuencia")
  #ax1.set_grid(True)
  ax1.legend()

  #Segundo subplot
  ax2.boxplot(data, patch_artist=True, boxprops=dict(facecolor='skyblue', color='black'), medianprops=dict(color='red'), whiskerprops=dict(color='black'), capprops=dict(color='black'))
  ax2.set_ylabel("Valores")
  #ax2.set_grid(True)
  ax2.legend()

  # Título centrado para toda la figura
  fig.suptitle(titulo, fontweight='bold', fontsize=16, y=1.05)

  # Ajustar el diseño
  plt.tight_layout()
  plt.show()

graficos_conjuntos(df['Total Amount'], 5, "Distribución por monto total de compra")

graficos_conjuntos(df['Quantity'], 4, "Distribución por cantidad de artículos por compra")
```

Es posible observar que, en el caso de la distribución por monto total de la compra, la cantidad de compras va disminuyendo drásticamente a a medida que el monto de la compra aumenta. Es decir, hay una gran concentración de compras de montos totales más bajos. Esto se refleja en el boxplot muy apretado hacia la parte inferior, casi sin bigote en esta parte, y una mediana de los datos muy similar al primer cuartil dando cuenta de esta concentración

La variación para la cantidad de artículos por compra presenta una evidente variación menor, influída por cierto por un menor rango de las mismas. De hecho, se trata de una distribución uniforme, en la práctica. Esto se refleja en el boxplot sin bigotes, con una mediana desplazada al valor inferior (primer cuartil) pero que debemos interpretar desde el hecho que, ante un rango muy pequeño de valores, no corresponde a un indicador tan significativo.

* Crea gráficos de líneas para mostrar las tendencias de ventas a lo largo del tiempo.

Vamos a crear los gráficos considerando por cierto la columna 'Date', y su relación con los montos totales de las ventas

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Total Amount', color='skyblue', errorbar=None)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

No se aprecia un patrón de variación de las ventas en el tiempo.

Vamos a construir el gráfico considerando la cantidad de artículos por compra.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Quantity', color='skyblue', errorbar=None)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

Nuevamente, tampoco se observa una tendencia o patrón

Utilizamos el comando **pairplot()** para graficar todas las variables, de a pares, para analizar posibles correlaciones.

```python
sns.pairplot(df)
```

Es posible observar que, salvo para las variabes que ex profeso estań relacionadas una con otra, no existen correlaciones apreciables entre los pares de variables. Construiremos el diagrama de puntos para las variables Total Amount y Age, separando por género

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.set_facecolor('lightgray')

sns.scatterplot(data=df, x='Age', y='Total Amount',hue='Gender', s=100, ax=ax)
ax.set_title('Relación entre Total Amount y Age, por Género', fontsize=20)
ax.set_xlabel('Edad', fontsize=15)
ax.set_ylabel('Total', fontsize=15)
ax.legend(loc='upper right', fontsize='large', shadow=False,facecolor='white')
```

No se obvserva correlación.
